```julia
using CSV
using DataFrames
using Plots
using Statistics
input="dane.csv"
df=CSV.read(input, delim=";")
```

```julia
df2=by(df, [:size, :type]) do subGroup
          DataFrame(mean = mean(subGroup[:time]), std = std(subGroup[:time])) 
        end
```

```julia
plot(df2[:size],  df2[:mean], group=df2[:type], layout=2, xlabel = "rozmiar wektora", ylabel = "czas w ns", colour = [:red :blue])
```

```julia
scatter(df2[:size], df2[:mean], group=df2[:type], layout=2, yerr=df2[:std],xlabel = "rozmiar wektora", ylabel = "czas w ns", colour = [:red :blue])
```
